```python
import pandas as pd
pd.options.display.max_rows = 100
```

```python
df_hh = pd.read_excel("../data/raw/ham_kita_suburb_updated.xls", index_col = [0])
df_hh = df_hh.rename(columns={"area": "stadtteil"})
```

```python
df_kinder = pd.read_excel("../data/raw/kinder_hh_stadtteile.xlsx")
df_kinder
```

```python
sum_area_kids = df_kinder.groupby(df_kinder["area"])["under6"].sum()
df_area_kids = sum_area_kids.reset_index()
df_area_kids
```

```python
sum_area_spots = df_hh.groupby(df_hh["stadtteil"])["kita_spots"].sum()
df_area_spots = sum_area_spots.reset_index()
df_area_spots
```

```python
df_area_spots["stadtteil"] = df_area_spots['stadtteil'].str.replace('Allermoehe', 'Allermöhe', regex=False)
```

```python
df_area_spots["stadtteil"] = df_area_spots["stadtteil"].str.replace("ae","ä",regex=False)
df_area_spots["stadtteil"] = df_area_spots["stadtteil"].str.replace("ue","ü",regex=False)
df_area_spots["stadtteil"] = df_area_spots["stadtteil"].str.replace("oe","ö",regex=False)
df_area_spots["stadtteil"] = df_area_spots["stadtteil"].str.replace("ss","ß",regex=False)
df_area_spots["stadtteil"] = df_area_spots["stadtteil"].str.replace("Neünfelde","Neuenfelde",regex=False)
df_area_spots["stadtteil"] = df_area_spots["stadtteil"].str.replace("Neüngamme","Neuengamme",regex=False)
```

```python
df_merge = pd.merge(df_area_spots, df_kinder, on="stadtteil", how="left")
df_merge
```

```python
sum_merge_area = df_merge.groupby(df_merge["area"])["kita_spots", "under6"].sum()
df_merge2 = sum_merge_area.reset_index()
df_merge2
```

```python
df_merge2["diff"] = df_merge2.under6-df_merge2.kita_spots
df_merge2["kids_per_spot"] = df_merge2.under6/df_merge2.kita_spots
#df_sums_all = df_sums_all.dropna()
df_merge2
```

```python
df_merge2["prct_missing"] = df_merge2["diff"]/df_merge2["under6"]
```

```python
df_merge["diff"] = df_merge.under6-df_merge.kita_spots
df_merge["kids_per_spot"] = df_merge.under6/df_merge.kita_spots
#df_sums_all = df_sums_all.dropna()
df_merge
```

```python
f_labor_force = [52.8, 56.7, 58.1, 52.0, 61.7, 52.0, 56.1]
```

```python
df_merge2["women_labor_percent"]=f_labor_force
```

```python
df_merge["prct_missing"] = df_merge["diff"]/df_merge["under6"]
```

```python
df_merge.to_csv("../data/derived/calc_data_hh.csv")
```
