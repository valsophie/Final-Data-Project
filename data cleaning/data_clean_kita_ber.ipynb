```python
import pandas as pd
```

```python
df_ber = pd.read_csv("../data/derived/kita_data_ber.csv", index_col=[0])
```

```python
df_ber.head(100)
```

```python
kita_list = list(df_ber.kita_address)
```

```python
plz = [x[:5] for x in kita_list]

df_ber["plz"] = plz
```

```python
df_ber
```

```python
df_ber.loc[df_ber['plz'] == "13187"]
```

```python
import numpy as np

df_ber["kita_spots"] = np.where(df_ber["kita_spots"].str.isdigit(), df_ber["kita_spots"], 0)
df_ber["plz"] = np.where(df_ber["plz"].str.isdigit(), df_ber["plz"], 0)
```

```python
df_ber["kita_spots"] = df_ber["kita_spots"].astype(int)
df_ber["plz"] = df_ber["plz"].astype(int)
```

```python
df_ber.to_csv(r'../data/derived/kita_data_berlin_clean.csv')
df_ber.to_json(r'../data/derived/kita_data_berlin_clean.json')
```
