```python
import pandas as pd
from pandas import DataFrame
import re
import datetime
import requests
import bs4
import csv
from bs4 import BeautifulSoup, NavigableString, Tag
```

```python
kita_name = []
kita_address = []
kita_spots = []
postal_code= [20097, 20099, 20144, 20146, 20148, 20149, 20249, 20251, 20253,
              20255, 20257, 20259, 20354, 20355, 20357, 20359, 20457, 20459, 20535, 20537,
              20539, 21029, 21031, 21033, 21035, 21037, 21039, 21073, 21075, 21077, 21079,
              21107, 21109, 21129, 21147, 21149, 22041, 22043, 22045, 22047, 22049, 22081,
              22083, 22085, 22087, 22089, 22111, 22115, 22117, 22119, 22143, 22145, 22147,
              22149, 22159, 22175, 22177, 22179, 22297, 22299, 22301, 22303, 22305, 22307,
              22309, 22335, 22337, 22339, 22359, 22391, 22393, 22395, 22397, 22399, 22415,
              22417, 22419, 22453, 22455, 22457, 22459, 22523, 22525, 22527, 22529, 22547,
              22549, 22559, 22587, 22589, 22605, 22607, 22609, 22761, 22763, 22765, 22767,
              22769]

for code in postal_code:
    super_spots = []
    current_page = 1
    while True: 
        page_kita=[]
        hola=f'https://www.kitanetz.de/kitasuche/kita-suchergebnis.php?go={current_page}&t_suchwort=&t_plz={code}&t_stadt=&t_land='
        print(hola)
        r = requests.get(hola)
        soup= BeautifulSoup(r.content, "html.parser")
        #print(soup)
        
        kitas2 = soup.find('div', class_="nocopy")
        
        k_name = kitas2.find_all('b')
        k_name = k_name[2:]
        k_name = [k.text for k in k_name]
        #print(k_name)
        #text = str(k_name).strip()
        kita_name.append(k_name)
        #print(kita_name)
        
        for br in kitas2.find_all('br'):
            address = br.nextSibling
            
            if not (address and isinstance(address,NavigableString)):
                continue
            
            address2 = address.nextSibling
            if address2 and isinstance(address2,Tag) and address2.name == 'br':
                text = str(address).strip()
                #address = address[1:]
                if text:
                    kita_address.append(address)
            #print(address)
            
            r = re.compile(r'^PLZ')
            filtered = [i for i in kita_address if not r.match(i)]
            kita_address = filtered
            
            kita_address = [k.replace("\nHamburg, ", "") for k in kita_address]
        
        
        x=0
        for br in kitas2.find_all("br"):
            
            spots = br.nextSibling
            
            clean_spot=str(spots)
            clean_spot=clean_spot.replace('<div class="slabel">','')
            clean_spot=clean_spot.split('\xa0')[0]
            if clean_spot == "Stichworte:":
                continue
            else:
                if not (spots and isinstance(spots,NavigableString)):
                    continue
                
                test_spot=spots.nextSibling.nextSibling
            
                test_spot=str(test_spot)
                test_spot=test_spot.replace('<div class="slabel">','')
                test_spot=test_spot.split('\xa0')[0]
                #test_spot=test_spot.replace('Es wurden','')
                #test_spot=test_spot.replace('Kitas gefunden','')
               # test_spot=[t.replace('Es wurden', '') for t in test_spot]
                
                if test_spot == "Alter:":
                    spots2 = spots.nextSibling.nextSibling.nextSibling.nextSibling.nextSibling
                    page_kita.append(spots2)
                    
                elif test_spot == "Plätze:":
                    spots3=spots.nextSibling.nextSibling.nextSibling
                    page_kita.append(spots3)
                else:
                    page_kita.append(0)     
        super_spots.append(page_kita[2:])
            
        pages = kitas2.find_all("span", class_="pages")[0].text
        pages2 = pages.split()
        #current_page  = int(pages2[1])
        last_page = int(pages2[3])
        print(last_page)
        if current_page >= last_page:
            break
        current_page+=1
            
    for spot in super_spots:
        for x in spot:
            kita_spots.append(x)
```

```python
test = list(zip(kita_name, kita_address, kita_spots))
test
```
