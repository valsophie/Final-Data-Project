```python
import pandas as pd
from pandas import DataFrame
import re
import datetime
import requests
import bs4
import csv
from bs4 import BeautifulSoup, NavigableString, Tag
```

```python
kita_name = []
kita_address = []
kita_spots = []
postal_code= [10115, 10117, 10119, 10178, 10179, 10243, 10245, 10247, 10249, 10315, 10317, 10318,
              10319, 10365, 10367, 10369, 10405, 10407, 10409, 10435, 10437, 10439, 10551, 10553,
              10555, 10557, 10559, 10585, 10587, 10589, 10623, 10625, 10627, 10629, 10707, 10709,
              10711, 10713, 10715, 10717, 10719, 10777, 10779, 10781, 10783, 10785, 10787, 10789,
              10823, 10825, 10827, 10829, 10961, 10963, 10965, 10967, 10969, 10997, 10999, 12043,
              12045, 12047, 12049, 12051, 12053, 12055, 12057, 12059, 12099, 12101, 12103, 12105,
              12107, 12109, 12157, 12159, 12161, 12163, 12165, 12167, 12169, 12203, 12205, 12207,
              12209, 12247, 12249, 12277, 12279, 12305, 12307, 12309, 12347, 12349, 12351, 12353,
              12355, 12357, 12359, 12435, 12437, 12439, 12459, 12487, 12489, 12524, 12526, 12527,
              12529, 12555, 12557, 12559, 12587, 12589, 12619, 12621, 12623, 12627, 12629, 12679,
              12681, 12683, 12685, 12687, 12689, 13051, 13053, 13055, 13057, 13059, 13086, 13088,
              13089, 13125, 13127, 13129, 13156, 13158, 13159, 13187, 13189, 13347, 13349, 13351,
              13353, 13355, 13357, 13359, 13403, 13405, 13407, 13409, 13435, 13437, 13439, 13465,
              13467, 13469, 13503, 13505, 13507, 13509, 13581, 13583, 13585, 13587, 13589, 13591,
              13593, 13595, 13597, 13599, 13627, 13629, 14050, 14052, 14053, 14055, 14057, 14059,
              14089, 14109, 14129, 14163, 14165, 14167, 14169, 14193, 14195, 14197, 14199]
for code in postal_code:
    super_spots = []
    current_page = 1
    while True: 
        page_kita=[]
        hola=f'https://www.kitanetz.de/kitasuche/kita-suchergebnis.php?go={current_page}&t_suchwort=&t_plz={code}&t_stadt=&t_land='
        print(hola)
        r = requests.get(hola)
        soup= BeautifulSoup(r.content, "html.parser")
        #print(soup)
        
        kitas2 = soup.find('div', class_="nocopy")
        
        k_name = kitas2.find_all('b')
        k_name = k_name[2:]
        k_name = [k.text for k in k_name]
        #print(k_name)
        #text = str(k_name).strip()
        kita_name.append(k_name)
        #print(kita_name)
        
        for br in kitas2.find_all('br'):
            address = br.nextSibling
            
            if not (address and isinstance(address,NavigableString)):
                continue
            
            address2 = address.nextSibling
            if address2 and isinstance(address2,Tag) and address2.name == 'br':
                text = str(address).strip()
                #address = address[1:]
                if text:
                    kita_address.append(address)
            #print(address)
            
            r = re.compile(r'^PLZ')
            filtered = [i for i in kita_address if not r.match(i)]
            kita_address = filtered
            
            kita_address = [k.replace("\nBerlin, ", "") for k in kita_address]
        
        
        x=0
        for br in kitas2.find_all("br"):
            
            spots = br.nextSibling
            
            clean_spot=str(spots)
            clean_spot=clean_spot.replace('<div class="slabel">','')
            clean_spot=clean_spot.split('\xa0')[0]
            if clean_spot == "Stichworte:":
                continue
            else:
                if not (spots and isinstance(spots,NavigableString)):
                    continue
                
                test_spot=spots.nextSibling.nextSibling
            
                test_spot=str(test_spot)
                test_spot=test_spot.replace('<div class="slabel">','')
                test_spot=test_spot.split('\xa0')[0]
                #test_spot=test_spot.replace('Es wurden','')
                #test_spot=test_spot.replace('Kitas gefunden','')
               # test_spot=[t.replace('Es wurden', '') for t in test_spot]
                
                if test_spot == "Alter:":
                    spots2 = spots.nextSibling.nextSibling.nextSibling.nextSibling.nextSibling
                    page_kita.append(spots2)
                    
                elif test_spot == "Plätze:":
                    spots3=spots.nextSibling.nextSibling.nextSibling
                    page_kita.append(spots3)
                else:
                    page_kita.append(0)     
        super_spots.append(page_kita[2:])
            
        pages = kitas2.find_all("span", class_="pages")[0].text
        pages2 = pages.split()
        #current_page  = int(pages2[1])
        last_page = int(pages2[3])
        print(last_page)
        if current_page >= last_page:
            break
        current_page+=1
            
    for spot in super_spots:
        for x in spot:
            kita_spots.append(x)
```
